```python
# waste_classifier_inference.ipynb
# Минимальный рабочий ноутбук для отправки на сайт

# ----------------------
# 1️⃣ Импорт библиотек
# ----------------------
import numpy as np
from tensorflow.keras.models import load_model
from PIL import Image

# ----------------------
# 2️⃣ Загружаем модель
# ----------------------
model = load_model('waste_classifier_efficientnet.h5')

# Классы (те же, что при обучении)
CLASSES = ['battery','biological','cardboard','clothes','glass',
           'metal','paper','plastic','shoes','trash']

# ----------------------
# 3️⃣ Функция предсказания
# ----------------------
def predict_category(img_path):
    # Загружаем и масштабируем изображение
    img = Image.open(img_path).convert('RGB')
    img_resized = img.resize((224, 224))
    x = np.array(img_resized)/255.0
    x = np.expand_dims(x, axis=0)

    # Предсказание
    preds = model.predict(x)[0]
    top_idx = np.argmax(preds)
    category = CLASSES[top_idx]
    confidence = preds[top_idx]*100

    print(f"Изображение относится к категории: **{category.upper()}**")
    print(f"Уверенность модели: {confidence:.2f}%")

# ----------------------
# 4️⃣ Пример использования
# ----------------------
img_path = 'glass_354.jpg'  # укажи путь к изображению для проверки
predict_category(img_path)
```
